```python
import pandas as pd
import numpy as np
```

```python
sc = pd.read_excel('SC_Community.xlsx', 'Sheet1')
ind = pd.read_excel('Indicators for Web Map.xlsx', 'Sheet1')
```

```python
df = sc[['Relevant Community', 'question_id', 'Convert To', 'question', 'response']].merge(ind[['Sector', 'question_name']], left_on='question_id', right_on='question_name', how='inner').drop('question_name', axis=1).sort_values(by=['question_id','Relevant Community'])
```

```python
# res = df.groupby(['question', 'response', 'Relevant Community']).size().reset_index(name='Freq')
# res = df.groupby(['question', 'response'], as_index=False)

qw = df.groupby(['question', 'response']).size().reset_index().rename(columns={0:'count'})
```

```python
writer = pd.ExcelWriter('SCandIndicator.xlsx')
qw.to_excel(writer, sheet_name='Sheet1', index=False)
writer.save()
```
