```python
import pandas as pd
```

```python
aba = pd.read_excel('REACH_ABA_Final_Datasets.xlsx', 'Administration_KI')
adm = pd.read_excel('Administration.xlsx', 'Fields_Tr')
choices = pd.read_excel('Administration.xlsx', 'choices')
survey = pd.read_excel('Administration.xlsx', 'survey')
```

```python
adm.head()
```

```python
questions = survey[survey.keep == 1][['name','type','label::English']]
```

```python
questions.head()
```

```python
## index table
index = adm[['Column Name', 'df_index','question_name']]
```

```python
index.head()
```

```python
numbers = pd.Series(range(0, len(aba.columns)))
aba.columns = numbers
```

```python
aba.head()
```

```python
id_vars_index = index['Column Name'].isin(['_id', 'KOATUU', 'Relevant Settlement', 'Raion_Name', 
                                      'Raion_KOATUU', 'Oblast_Name', 'Oblast_KOATUU', 'BSU ID',
                                      'Include', 'Admin2', 'Admin2_Name', 'Admin1', 'Admin1_Name'])

data_unpivoted = pd.melt(aba, id_vars=list(index['df_index'][id_vars_index]))

# data_unpivoted = pd.melt(aba, id_vars=index['df_index'][index['Column Name'].isin(['_id', 'KOATUU', 'Relevant Settlement', 'Raion_Name', 
#                                       'Raion_KOATUU', 'Oblast_Name', 'Oblast_KOATUU', 'BSU ID',
#                                       'Include', 'Admin2', 'Admin2_Name', 'Admin1', 'Admin1_Name'])])
```

```python
data_unpivoted.head()
```

```python
data_unpivoted = data_unpivoted.merge(adm, left_on='variable', 
                    right_on = 'df_index', how='left')
```

```python
data_unpivoted
```

```python
names = pd.Series(index['Column Name'][index.df_index.isin(data_unpivoted.columns[pd.Series(data_unpivoted.columns.values).str.isnumeric() != False])])
names = names.append(pd.Series(data_unpivoted.columns[pd.Series(data_unpivoted.columns.values).str.isnumeric() == False])).reset_index(drop=True)
data_unpivoted.columns = names
data_unpivoted.head()
```

```python
choices.head()
```

```python
# res_ = data_unpivoted.merge(choices[['name', 'label::English']].drop_duplicates(subset=['name', 'label::English']), 
#                            left_on='option_name', 
#                            right_on='name', how='left').sort_values('_id').reset_index(drop=True)
```

```python
res = data_unpivoted.merge(choices, 
                           left_on=['option_name','list_name'], 
                           right_on=['name','list_name'], how='left').sort_values('_id').reset_index(drop=True)
```

```python
res.head()
```

```python
res.drop(['name', 'label::Russian', 'Oblast'],axis=1, inplace=True)
```

```python
# П.С. choices[['name', 'label::English']].drop_duplicates(subset=['name', 'label::English']) такой же размер рез матр
# .drop('df_index', axis=1, inplace=True)

res.rename(columns={'Column Name': 'column_name', 'label::English': 'option_label'}, inplace=True)
# res['question_label'] = res['question_label'].str.replace('\?/', '*')
# res['question_label'] = res['question_label'].str.split('*', expand=True)
res
```

```python
res = res.merge(questions, left_on='question_name', right_on='name', how='left').drop(['name','type_y'],axis=1).sort_values(['_id','Position']).reset_index(drop=True)
```

```python
res.rename(columns={'type_x': 'type', 'label::English': 'question_label'}, inplace=True)
```

```python
res['options'] = ''
# res.loc[(res.type == 'select_multiple') & ~(res.option_label == np.nan), 'options'] = 1
res.loc[(res.type == 'select_multiple') & pd.notnull(res.option_label), 'options'] = 1
```

```python
writer = pd.ExcelWriter('Result.xlsx')
res.to_excel(writer, sheet_name='Administration', index=False)
writer.save()
```
